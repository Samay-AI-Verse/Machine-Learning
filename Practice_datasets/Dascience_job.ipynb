```python
from sklearn.preprocessing import LabelEncoder
```

```python
obj = LabelEncoder()
```

```python
newdf['gender_decoder'] = obj.fit_transform(newdf['gender'])
```

```python
newdf['company_type_decoder'] = obj.fit_transform(newdf['company_type'])
```

```python
newdf.head()
```

```python
x = newdf[['gender_decoder','experience']]
y = newdf['company_type_decoder']
```

```python
dict = {
    'decodercode':[5, 1, 0, 3, 4, 2, 6],
    'name':['Pvt Ltd', 'Funded Startup', 'Early Stage Startup', 'Other',
       'Public Sector', 'NGO', 'nan']
}
```

```python
dataframe = pd.DataFrame(dict)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[1,20]])
```
