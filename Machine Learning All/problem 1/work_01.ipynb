# Model Building actualy doing here 

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
df.head()
category = ["education","self_employed"]
scaller = ["income_annum","loan_amount","residential_assets_value","commercial_assets_value","luxury_assets_value","bank_asset_value"]
normal = ["no_of_dependents","loan_term","cibil_score"]
```

```python
preprocesser = ColumnTransformer(
    transformers= [
        ("cat", OneHotEncoder(handle_unknown="ignore"), category),
        ("scaller", StandardScaler(),scaller),
        ("normal", "passthrough",normal)
    ]
)
```

```python
pipe = Pipeline(
    [
        ("preprocesser", preprocesser),
        ("model",RandomForestClassifier(random_state=42))
    ]
)
```

```python
RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper_para = {
    "model__criterion":['gini', 'entropy', 'log_loss'],
    "model__max_depth":[3,5,10],
    "model__n_estimators":[100,200,300],
    "model__min_samples_split":[2,5,10]
}
```

```python
grid = GridSearchCV(
    estimator=pipe,
    param_grid=hyper_para,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_pipeline = grid.best_estimator_
```
