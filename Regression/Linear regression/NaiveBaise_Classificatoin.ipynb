# Classification algorithamsn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv('placement.csv')
```

```python
data = data.iloc[:,1:]
```

```python
data.head()
```

```python
plt.figure(figsize = (4,3))
sns.scatterplot(x = 'cgpa',y = 'iq',data = data,hue = 'placement')
plt.show()
```

```python
sns.kdeplot(data = data['iq'])
```

```python
x = data.iloc[:,:-1]
y = data['placement']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[6.8,123.0]])
```

```python
plt.figure(figsize = (5,4))
plot_decision_regions(x_train.to_numpy(),y_train.to_numpy(),clf= lr)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
ga = GaussianNB()
```

```python
ga.fit(x_train,y_train)
```

```python
ga.score(x_train,y_train)*100
```

```python
ga.score(x_test,y_test)*100
```

```python
plt.figure(figsize = (5,4))
plot_decision_regions(x_train.to_numpy(),y_train.to_numpy(),clf= ga)
plt.show()
```

```python
ml = MultinomialNB()
```

```python
ml.fit(x_train,y_train)
```

```python
ml.score(x_test,y_test)
```

```python
plt.figure(figsize = (5,4))
plot_decision_regions(x_train.to_numpy(),y_train.to_numpy(),clf= ml)
plt.show()
```

```python
be = BernoulliNB()
```

```python
be.fit(x_train,y_train)
```

```python
be.score(x_test,y_test)*100,be.score(x_train,y_train)*100
```

```python
plt.figure(figsize = (5,4))
plot_decision_regions(x_train.to_numpy(),y_train.to_numpy(),clf= be)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2, include_bias=False)
```

```python
x_train_poly = pf.fit_transform(x_train)
```

```python
x_test_poly = pf.transform(x_test)
```

```python
lr.fit(x_train_poly,y_train)
```

```python
lr.score(x_test_poly,y_test)*100
```
