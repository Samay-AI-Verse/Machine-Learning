```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Housing.csv')
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
data['mainroad'] = label.fit_transform(data['mainroad'])
```

```python
data['hotwaterheating'] = label.fit_transform(data['hotwaterheating'])
data['airconditioning'] = label.fit_transform(data['airconditioning'])
data['prefarea'] = label.fit_transform(data['prefarea'])
data['furnishingstatus'] = label.fit_transform(data['furnishingstatus'])
```

```python
data['guestroom'] = label.fit_transform(data['guestroom'])
data['basement'] = label.fit_transform(data['basement'])
```

```python
# sns.pairplot(data = data.corr())
```

```python
plt.figure(figsize = (12,5))
sns.heatmap(data = data.corr(),annot=True)
plt.show()
```

```python
x = data.iloc[:,1:]
list = [x.columns]
```

```python
x
```

```python
y = data.iloc[:,:1]
```

```python
sc = StandardScaler()
```

```python
sc.fit(x)
```

```python
x = pd.DataFrame(sc.transform(x),columns = x.columns)
```

# Linear Regression

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
objR = LinearRegression()
```

```python
objR.fit(x_train,y_train)
```

```python
objR.score(x_test,y_test)*100
```

```python
arr = objR.coef_[0]
```

```python
plt.figure(figsize = (18,6))
plt.bar(x.columns,arr)
plt.title('Linear Regression')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.show()
```

```python
x.columns
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

# Lasso

```python
la = Lasso(alpha=0.01)
```

```python
la.fit(x_train,y_train)
```

```python
la.score(x_test,y_test)
```

```python
plt.figure(figsize = (18,6))
plt.bar(x.columns,la.coef_)
plt.title('Lasso')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.show()
```

# Ridge

```python
ra = Ridge(alpha= 10)
```

```python
ra.fit(x_train,y_train)
```

```python
ra.score(x_test,y_test)*100
```

```python
ra.coef_
```

```python
plt.figure(figsize = (18,6))
plt.bar(x.columns,ra.coef_[0])
plt.title('Ridge')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.show()
```
