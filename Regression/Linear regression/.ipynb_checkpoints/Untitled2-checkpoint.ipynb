```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.pipeline import make_pipeline

# Load dataset
file_path = "Housing.csv"
df = pd.read_csv(file_path)

# Function to remove outliers using IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers from 'area' and 'price'
df_cleaned = remove_outliers(df, 'area')
df_cleaned = remove_outliers(df_cleaned, 'price')

# Convert categorical variables into numerical (One-Hot Encoding)
df_cleaned_encoded = pd.get_dummies(df_cleaned, columns=['mainroad', 'guestroom', 'basement', 'hotwaterheating',
                                                         'airconditioning', 'prefarea', 'furnishingstatus'], drop_first=True)

# Selecting relevant features
X = df_cleaned_encoded.drop(columns=['price'])  # All columns except price
y = df_cleaned_encoded['price']

# Create polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Split the data
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y, test_size=0.2, random_state=42)

# Train the polynomial regression model
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train_poly)

# Predict on test set
y_pred_poly = poly_model.predict(X_test_poly)

# Evaluate the model
r2_poly = poly_model.score(X_test_poly, y_test_poly) * 100
print(f"Polynomial Regression Accuracy: {r2_poly:.2f}%")
```
