```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
```

```python
data = data.iloc[:,2:]
```

```python
data
```

```python
data['Purchased'].value_counts()
```

```python
x = data.iloc[:,:-1]
```

```python
y = data['Purchased']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lo = LogisticRegression()
```

```python
lo.fit(x_train,y_train)
```

```python
lo.score(x_test,y_test)*100
```

```python
lo.predict([[46	,41000]])
```

```python
import imblearn
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rm = RandomUnderSampler()
```

```python
rm_x,rm_y = rm.fit_resample(x,y)
```

```python
rm_y.value_counts()
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(rm_x,rm_y,test_size=0.2,random_state=42)
```

```python
lo1 = LogisticRegression()
```

```python
lo1.fit(x_train1,y_train1)
```

```python
lo1.score(x_test1,y_test1)
```

```python
lo1.predict([[35,20000]])
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro = RandomOverSampler()
```

```python
ro_x,ro_y = ro.fit_resample(x,y)
```

```python
x_train2,x_test2,y_train2,y_test2 = train_test_split(ro_x,ro_y,test_size=0.2,random_state=42)
```

```python
lo2 = LogisticRegression()
```

```python
lo2.fit(x_train2,y_train2)
```

```python
lo2.score(x_test2,y_test2)
```
