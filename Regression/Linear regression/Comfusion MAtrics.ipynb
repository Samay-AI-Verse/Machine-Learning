```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('placement.csv')
```

```python
data = data.iloc[:,1:]
```

```python
data.head()
```

```python
plt.figure(figsize = (6,4))

sns.scatterplot(x = 'cgpa',y = 'iq',data = data,hue = 'placement')

plt.show()
```

```python
x = data.iloc[:,:-1]
y = data[['placement']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
obj = PolynomialFeatures()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
obj = LogisticRegression()
```

```python
obj.fit(x_train,y_train)
```

```python
obj.score(x_test,y_test)*100
```

```python
obj.predict([[6.8	,123.0]])
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
```

```python
cf = confusion_matrix(y_test,obj.predict(x_test))
```

```python
cf = confusion_matrix(y_test,obj.predict(x_test))
```

```python
sns.heatmap(cf,annot=True,fmt='d',cmap='Blues')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, obj.predict(x_test))
precision = precision_score(y_test, obj.predict(x_test))
recall = recall_score(y_test, obj.predict(x_test))
f1 = f1_score(y_test, obj.predict(x_test))

print(f"Accuracy: {accuracy *100:.2f}")
print(f"Precision: {precision*100:.2f}")
print(f"Recall: {recall*100:.2f}")
print(f"F1 Score: {f1*100:.2f}")
```
