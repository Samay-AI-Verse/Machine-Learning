```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Ice_cream selling data.csv')
```

```python
data
```

```python
x = data[['Temperature (°C)']]
y = data['Ice Cream Sales (units)']
```

```python
plt.figure(figsize = (6,3))
plt.title('polynominal regression')
plt.scatter(data['Temperature (°C)'],data['Ice Cream Sales (units)'])
plt.plot()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
obj = PolynomialFeatures(degree=2)
```

```python
obj.fit(x)
```

```python
x = obj.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
li = LinearRegression()
```

```python
li.fit(x_train,y_train)
```

```python
li.score(x_test,y_test)
```

```python
pred = li.predict(x)
```

```python
plt.figure(figsize = (6,3))
plt.title('polynominal regression')
plt.scatter(data['Temperature (°C)'],data['Ice Cream Sales (units)'])
plt.plot(data['Temperature (°C)'], pred,c = 'r')
plt.legend(['Original','Predict Line'])
plt.show()
```
