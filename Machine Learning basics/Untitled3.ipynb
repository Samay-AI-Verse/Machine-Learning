```python
import matplotlib.pyplot as plt

# Features
x = df['Glucose']
y = df['BMI']

# Target
labels = df['Outcome']

plt.scatter(x, y, c=labels)
plt.xlabel('Glucose')
plt.ylabel('BMI')
plt.title('Glucose vs BMI (colored by Outcome)')
plt.show()
```
